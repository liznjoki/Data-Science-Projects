###Continent

```python
import pycountry_convert as pc

# Function to convert country to continent
def convert_country_to_continent(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        return pc.convert_continent_code_to_continent_name(continent_code)
    except:
        return 'Not found'

# Apply the function to the DataFrame
df['continent'] = df['country'].apply(convert_country_to_continent)

# Show the resulting DataFrame
df.head(5)
```

```python
# Group 'col2' values by 'col1' and get the number of unique values in each group
grouped = df.groupby('continent')['country'].nunique()
grouped
# Reset the index to make 'col1' and 'col2' columns
grouped = grouped.reset_index(name='count')
grouped.sort_values(by="count", ascending=False)

# Show the resulting DataFrame
```
